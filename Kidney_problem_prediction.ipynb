```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

# feature scalling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test  = sc.fit_transform(X_test)
X_train
```

# model selection

```python
def models(X_train,Y_train):

 
 from sklearn.linear_model import LogisticRegression
 log = LogisticRegression(random_state=0)
 log.fit(X_train,Y_train)

 from sklearn.tree import DecisionTreeClassifier
 tree=DecisionTreeClassifier( criterion= 'entropy',random_state=0 )
 tree.fit(X_train,Y_train)

 from sklearn.ensemble import RandomForestClassifier
 forest=RandomForestClassifier(n_estimators=10,criterion= 'entropy' ,random_state=0)
 forest.fit(X_train,Y_train)


 print('[0] LinearRegressionear_model',log.score(X_train,Y_train))
 print('[1]  DecisionTreeClassifier_model',tree.score(X_train,Y_train))
 print('[2] RandomForestRegressor_model',forest.score(X_train,Y_train))
    
 return log, tree, forest 

model=models(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
 print(classification_report(Y_test , model[0].predict(X_test)))
 print(accuracy_score(Y_test , model[0].predict(X_test)))
```
